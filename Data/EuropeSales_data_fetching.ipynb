```python
#europe1['Total BEV'] = europe1['Total BEV(1)'].str.split('[').str[0]
#europe1['Total BEV(1)'] = europe1['Total BEV(1)'].astype(str)
```

```python
europe1 = europe1.rename(columns={"Total BEV(1)": "Total_BEV", "Total PHEV(2)": "Total_PHEV", 
                        "Total PEV(3)": "Total_PEV", "PEVmarket share(4)": "PEV_Marketshare",
                       "Growth(BEVs)": "BEV_Growth", "Growth(PHEVs)": "PHEV_Growth", "Growth(PEVs)":"PEV_Growth"})
```

```python
europe2 = europe1.values.tolist()
```

```python
e1 = pd.DataFrame(europe2, columns=['Year','Total_BEV','BEV_Growth','Total_PHEV','PHEV_Growth',
                                    'Total_PEV','PEV_Growth','PEV_Marketshare'])
```

```python
#e1.dtypes
e1 = e1.drop(columns=['BEV_Growth','PHEV_Growth','PEV_Marketshare','Total_PEV','PEV_Growth'])
```

```python
e1['Total_BEV'] = e1['Total_BEV'].str.split('[').str[0]
e1['Total_PHEV'] = e1['Total_PHEV'].str.split('[').str[0]
```

```python
#pd.to_numeric(e1['Total_BEV'], downcast='float')
e1.replace(',','', regex=True, inplace=True)
e1
```

```python
e1.to_csv('Clean_files/Europe1.csv', index=False)
```
